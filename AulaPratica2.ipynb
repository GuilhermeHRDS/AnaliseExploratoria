```python
# The UserWarning "No artists with labels found to put in legend" indicates that `plt.legend()`
# was called without explicit labels or when the legend was already handled by seaborn's `hue` parameter.
# To fix this, remove the `plt.legend(loc="center")` line from your plotting cells (qWyOmR0vAhUU and M1WbSjZEBYGe).
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv("https://raw.githubusercontent.com/dayanebravo/AED/main/pes_2012.csv")
df1.head()
```

```python
df1["V4718"] = pd.to_numeric(df1["V4718"], errors="coerce")
df1["V4720"] = pd.to_numeric(df1["V4720"], errors="coerce")
```

```python
df = df1.dropna()
df.head()
```

```python
#Bibliotecas par agrafico

import matplotlib.pyplot as plt #Biblioteca para graficos
import seaborn as sn
```

```python
sn.set(style="whitegrid")
```

```python
#Grafico de Barras Horizontal
ax = sn.countplot(y=df.V0302, hue=df.V0302)
plt.title("Sexo")
plt.ylabel("")
plt.xlabel("")

# Para mostrar a legenda corretamente quando usamos o hue
plt.legend(title='Sexo', loc='center right', bbox_to_anchor=(1.25, 0.5))

plt.show()
```

```python
#Grafico de Barras Vertical
ax = sn.countplot(x=df.V0302, hue=df.V0302)
plt.title("Sexo")
plt.ylabel("")
plt.xlabel("")

# Ajustando a legenda para não sobrepor as barras
plt.legend(title='Sexo', loc='upper right')

plt.show()
```

```python
sn.countplot(x=df.V0302, hue=df.V0404)
plt.title("Sexo")
plt.ylabel("")
plt.xlabel("")
plt.legend(loc="center")

plt.legend(bbox_to_anchor=(1.3, 0.8))

plt.show()
```

```python
from matplotlib.patches import bbox_artist
# grafico de Sizes (Pizza)

sizes = df.V0404.value_counts()
h = df["V0404"].value_counts().index
plt.figure(figsize=(8,8))

plt.pie(sizes, labels=h, explode=(0,0,0.2,0.3,0.4), autopct='%1.2f%%',  startangle=90)
plt.legend(bbox_to_anchor=(1.3, 0.8))
plt.show()

```
