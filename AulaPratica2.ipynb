```python
# The UserWarning "No artists with labels found to put in legend" indicates that `plt.legend()`
# was called without explicit labels or when the legend was already handled by seaborn's `hue` parameter.
# To fix this, remove the `plt.legend(loc="center")` line from your plotting cells (qWyOmR0vAhUU and M1WbSjZEBYGe).
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv("https://raw.githubusercontent.com/dayanebravo/AED/main/pes_2012.csv")
df1.head()
```

```python
df1["V4718"] = pd.to_numeric(df1["V4718"], errors="coerce")
df1["V4720"] = pd.to_numeric(df1["V4720"], errors="coerce")
```

```python
df = df1.dropna()
df.head()
```

```python
#Bibliotecas par agrafico

import matplotlib.pyplot as plt #Biblioteca para graficos
import seaborn as sn
```

```python
sn.set(style="whitegrid")
```

```python
#Grafico de Barras

sn.countplot(y=df.V0302, hue=df.V0302) #grafico barra - frequencia da coluna indicada no eixo y
plt.title("Sexo")

plt.ylabel("")
plt.xlabel("")

#plt.legend(loc="center")

plt.show()
```

```python
sn.countplot(x=df.V0302, hue=df.V0302)
plt.title("Sexo")
plt.ylabel("")
plt.xlabel("")

#plt.legend(loc="center")

plt.show()
```
